TOTAL RUNS SCORED BY EACH TEAM ACROSS THE TOURNAMENT

```python
total_scores_per_team_df=df1.groupby('batting_team')['total_runs'].sum().reset_index()
total_scores_per_team_df.index+=1
teams=total_scores_per_team_df['batting_team'].values
max_runs=total_scores_per_team_df['total_runs'].values
total_scores_per_team_df
```

```python
plt.figure(figsize=(10,6))
c=['#F5D43D','#17449B','#0D243C','#3A225D','#1C4F9C','#005DA0','#D71920','#EA1A85','#DA1818','#F9A01B']
bars=plt.barh(teams,max_runs,color=c)
for bar,run in zip(bars,max_runs):
    plt.text(
        bar.get_width(),
        bar.get_height()-0.5+bar.get_y(),
        f'{(run)}',
                  va='center'
        )

plt.title('Total Runs Scored By Each Team')
plt.xlabel('Runs')
plt.ylabel('Teams')
plt.show()
```

```python
c2 = ['#17449B', '#228B22', '#187BCD', '#3A225D', '#4682B4', '#DA291C', '#FFD700', '#FF9933', '#E40046', '#FF8C00', '#EA1A85', '#0048BA']
plt.figure(figsize=(10, 6))
bars = plt.barh(stadium, runs, color=c2)

for bar, run in zip(bars, runs):
    plt.text(
        bar.get_width() + 0.5, 
        bar.get_y() + bar.get_height() / 2, 
        f'{run}', 
        va='center'
    )

plt.title('Runs scored by Venue')
plt.xlabel('Runs')
plt.ylabel('Stadium')
plt.show()
```

INNINGS PHASE ANALYSIS

```python
def innings_phase(over):
    if 0<= over <=6:
        return 'PowerPlay'
    elif 6< over <=15:
        return 'Middle Overs'
    else:
        return 'Death Overs'

df1['innings_phase']=df1['over'].apply(innings_phase)
phase_runs=df1.groupby(['innings','innings_phase'])['total_runs'].sum().unstack().fillna(0)


phase_runs.plot(kind='bar',stacked=True,color=['#1f77b4', '#ff7f0e', '#2ca02c'], figsize=(8, 6))
plt.title('Run Scoring Trends Across Innings Phases')
plt.xlabel('Innings')
plt.ylabel('Total Runs')
plt.legend(title='Innings Phase', loc='center right')
plt.show()
```

The above graph shows that most runs are scored in the order:: Middle Overs--> PowerPlay --> Death Overs
